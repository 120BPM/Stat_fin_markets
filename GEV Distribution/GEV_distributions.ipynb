```python

from scipy.stats import genextreme

from scipy.stats import weibull_min
import numpy as np
import matplotlib.pyplot as plt
mu=1
sigma=2
csi=0

fig, ax = plt.subplots(figsize=(20, 10))

x = np.linspace(genextreme.ppf(0.01, csi,mu,sigma),
                genextreme.ppf(0.99,csi,mu,sigma), 1000)
ax.plot(x, genextreme.pdf(x,csi,mu,sigma),
       'r-', lw=5, alpha=0.8, label='Gumbel pdf')
ax.legend(loc='best', frameon=False)

mu=1
sigma=2

csi=0.97
x = np.linspace(weibull_min.ppf(0.01, csi,mu,sigma),
              weibull_min.ppf(0.99, csi,mu,sigma), 1000)
plt.plot(x,weibull_min.pdf(x, csi,mu,sigma),
         'b-', lw=3, alpha=0.8, label='Frechet pdf')
ax.legend(loc='best', frameon=False)




mu=1
sigma=1

csi=-0.05
x = np.linspace(genextreme.ppf(0.01, csi,mu,sigma),
              genextreme.ppf(0.99, csi,mu,sigma), 1000)
plt.plot(-x, genextreme.pdf(-x, csi,mu,sigma),
         'g-', lw=3, alpha=0.8, label='Weibull pdf')
ax.legend(loc='best', frameon=False)

plt.show()
```
